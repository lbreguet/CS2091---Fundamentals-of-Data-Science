```python
import numpy as np
import pandas as pd
import math

data = pd.read_csv('https://raw.githubusercontent.com/AUP-CS2091/Week-2/master/titanic-data.csv')

# data is going to contain the modified data which contain numerical values
cleaned_data = data
cleaned_data['Sex'] = data['Sex'].astype('category').cat.codes

# Age sometimes contains nan (not a number) or -1 values, lets replace them by 0s
cleaned_data['Age'] = data['Age'].map(lambda x: 0 if math.isnan(x) or x == -1 else x)
```

## Classification Trees

```python
# Our feature
X = cleaned_data[['Sex','Pclass','Age']]

# Our target
Y = cleaned_data['Survived']

from sklearn import tree

clf = tree.DecisionTreeClassifier(criterion='entropy',max_depth=8)
clf = clf.fit(X, Y)
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=X.columns.values,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
graph
```

#### 4)

```python
import numpy as np
import pandas as pd
import math

data = pd.read_csv('./Video_Games_Sales_as_at_22_Dec_2016.csv')

# data is going to contain the modified data which contain numerical values
cleaned_data = data
cleaned_data['Publisher'] = data['Publisher'].astype('category').cat.codes

cleaned_data['Critic_Score'] = data['Critic_Score'].map(lambda x: 0 if math.isnan(x) or x == -1 else x)
cleaned_data['Critic_Count'] = data['Critic_Count'].map(lambda x: 0 if math.isnan(x) or x == -1 else x)
cleaned_data['User_Count'] = data['User_Count'].map(lambda x: 0 if math.isnan(x) or x == -1 else x)
cleaned_data
```

#### 5)

```python
# Our feature
X = cleaned_data[['Global_Sales','Critic_Score','User_Count']]

# Our target
Y = cleaned_data['Platform']

from sklearn import tree

clf = tree.DecisionTreeClassifier(criterion='entropy',max_depth=3)
clf = clf.fit(X, Y)
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=X.columns.values,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
graph
```

#### 6)

```python
# trying to discern whether a game did well globally

df = data.to_dict(orient='records')
d = df[16]

def predict(val):
    if val["Global_Sales"] > 20:
        if val["User_Score"] >= str(75):
            if val["Critic_Score"] > 80:
                return True
            else:
                return False
        else:
            return False
    else:
        return False
                
predict(d)
```
