```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/AUP-CS2091/Week-7/master/churn_train.csv?token=AFj8VRqjWJV3qHTNigar9MFsnq68-cFcks5Z8ie8wA%3D%3D')
tdata = pd.read_csv('https://raw.githubusercontent.com/AUP-CS2091/Week-7/master/churn_test.csv?token=AFj8VUJJdGcDbMzoty9Joi26M-XTPoOtks5Z8iftwA%3D%3D')
data.head()
```

* COLLEGE : Is the customer college educated?
* INCOME	: Annual income
* OVERAGE	: Average overcharges per month
* LEFTOVER : Average % leftover minutes per month
* HOUSE : Value of dwelling (from census tract)
* HANDSET_PRICE : Cost of phone
* OVER_15MINS_CALLS_PER_MONTH : Average number of long (>15 mins) calls per month
* AVERAGE_CALL_DURATION : Average call duration
* REPORTED_SATISFACTION : Reported level of satisfaction
* REPORTED_USAGE_LEVEL : Self-reported usage level
* CONSIDERING_CHANGE_OF_PLAN : Was customer considering changing his/her plan?
* LEAVE : Class variable: whether customer left or stayed

```python
cleaned_data = data
cleaned_data['COLLEGE'] = data['COLLEGE'].astype('category').cat.codes
cleaned_data['REPORTED_SATISFACTION'] = data['REPORTED_SATISFACTION'].astype('category').cat.codes
cleaned_data['REPORTED_USAGE_LEVEL'] = data['REPORTED_USAGE_LEVEL'].astype('category').cat.codes
cleaned_data['CONSIDERING_CHANGE_OF_PLAN'] = data['CONSIDERING_CHANGE_OF_PLAN'].astype('category').cat.codes
cleaned_data['LEAVE'] = data['LEAVE'].astype('category').cat.codes
cleaned_data.dropna()
cleaned_data.head()
```

```python
cleaned_tdata = tdata
cleaned_tdata['COLLEGE'] = tdata['COLLEGE'].astype('category').cat.codes
cleaned_tdata['REPORTED_SATISFACTION'] = tdata['REPORTED_SATISFACTION'].astype('category').cat.codes
cleaned_tdata['REPORTED_USAGE_LEVEL'] = tdata['REPORTED_USAGE_LEVEL'].astype('category').cat.codes
cleaned_tdata['CONSIDERING_CHANGE_OF_PLAN'] = tdata['CONSIDERING_CHANGE_OF_PLAN'].astype('category').cat.codes
cleaned_tdata.dropna()
cleaned_tdata.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

X = cleaned_data.loc[:, data.columns != 'LEAVE']
y = cleaned_data['LEAVE']

X_pred = cleaned_tdata

model = DecisionTreeClassifier(max_depth=10)
model.fit(X,y)
probs = model.predict_proba(X_pred)

cleaned_tdata['PLEAVE'] = probs[:,0]
tdata.head()
```
